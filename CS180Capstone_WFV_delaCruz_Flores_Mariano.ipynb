## Dimensional Reduction using Truncated SVD

Because the data is extremely sparse and high-dimensional, the dataset suffers from the curse of dimensionality. There are 2 approaches to clustering in this space.

One is to use subspace clustering methods to be able to create clusters in lower dimensions. However, no such methods can be found in scikit-learn that can easily be implemented.

Another is to use dimensionality reduction to transform the data into a lower dimension, where clustering can be reasonably performed. While easier to perform, methods like feature transformation maintain data from irrelevant dimensions (of which there are many in our dataset), and thus can maintain the noisiness of the dataset.

Despite these drawbacks, its ease of use urged us to use feature transformation, specifically Truncated SVD (which works well on sparse data), as a means to reduce the dimensionality of the dataset.

(Source: http://www.kdd.org/exploration_files/parsons.pdf)

```python
from sklearn.decomposition import TruncatedSVD

import matplotlib.pyplot as plt
```

A random seed is given to the TruncatedSVD class before running, for reproducibility's sake.

```python
svd = TruncatedSVD(n_components=200, n_iter=30, random_state=180)
U = svd.fit_transform(rating_fv)

n_samples = len(U)
n_features = svd.n_components
print("{} samples, {} features".format(n_samples, n_features))
np.shape(U)
```

200 components give a nice total explained variance ratio of ~55%.

```python
print("Explained variance percent: {}%".format(int(svd.explained_variance_ratio_.sum() * 10000) / 100))
```

```python
cumulative_evr = [int(svd.explained_variance_ratio_[0:i+1].sum() * 10000) / 100 for i in range(n_features)]
x_features = [i+1 for i in range(n_features)]
plt.plot(x_features, cumulative_evr)
plt.title("Cumulative explained variance ratio vs number of components")
plt.xlabel("Number of components (based on the first n eigenvalues)")
plt.ylabel("Cumulative e.v.r. (in percent)")
plt.show()
```

## Using KMeans score and Silhouette score to determine a good number of clusters

Part of the problem when it comes to clustering data is to figure out the best number of cluster centroids to use. Too little and the separations are too general to be useful. Too much and the separations are too specific to be reliable.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

### KMeans and Silhouette score calculation for 2 to 50 clusters

```python
clusters = []
k_score = []
silhouettes = []

base = 2
end = 50

for n_clusters in range(base,end+1):
    kmeans = KMeans(n_clusters=n_clusters, random_state=180)
    cluster_labels = kmeans.fit_predict(X)

    silhouette_avg = silhouette_score(X, cluster_labels)
    cluster_score = kmeans.score(X)

    clusters.append(n_clusters)
    k_score.append(cluster_score)
    silhouettes.append(silhouette_avg)
    
    print("{}...".format(n_clusters), end='', flush=True)
```

The graph for the KMeans score is a curve with an elbow point, where succeeding clusters starts to lose effectiveness and experience diminishing returns. This point determines the number of clusters to be used in the actual clustering of the dataset.

```python
plt.plot(clusters, k_score)
plt.title('K score')
plt.show()

plt.plot(clusters, silhouettes)
plt.title('Silhouette score')
plt.show()
```

Again, the arrays are persisted for the same reason as before.

```python
np.save("k_scores", np.asarray(k_score))
np.save("silhouette_scores", np.asarray(silhouettes))
```

### Finding the KMeans elbow

```python
from scipy.spatial.distance import cdist
```

First, the KMeans scores are standardized in a range of 0 to 50, so that the following computations become easy to work with.

```python
min_k = min(k_score)
k_score_std = k_score - min_k
max_k = max(k_score_std)
k_score_std = k_score_std * 50 / max_k
```

Next, the actual finding of the elbow is performed.

The elbow point of a curve is defined to be the point farthest from the line segment produced when connecting the first and last points. Mathematically speaking, this point has the largest magnitude of its projection to the line segment's normal vector, assuming the first point lies in the origin.

Using the following computations, the point at which the elbow occurs is determined to be at 13 clusters.

Source: https://dataplatform.ibm.com/analytics/notebooks/54d79c2a-f155-40ec-93ec-ed05b58afa39/view?access_token=6d8ec910cf2a1b3901c721fcb94638563cd646fe14400fecbb76cea6aaae2fb1

```python
lenx = len(clusters)

k_curve = np.vstack((clusters, k_score_std)).T

first = k_curve[0]
line_vec = k_curve[-1] - first
line_norm = line_vec / cdist([line_vec], [np.zeros(2)], 'euclidean')[0][0]

vec_from_first = k_curve - first

scalar_prod = np.sum(vec_from_first * np.matlib.repmat(line_norm, lenx, 1), axis=1)
vec_from_first_x = np.outer(scalar_prod, line_norm)

vec_to_line = vec_from_first - vec_from_first_x

dists = cdist(vec_to_line, np.zeros((lenx,2)), 'euclidean')
dists_to_line = np.diagonal(dists)

elbow = np.argmax(dists_to_line)
K_elbow = clusters[elbow]
print("Number of clusters to use: {}".format(clusters[elbow]))
```

The KMeans score and Silhouette score at this number of clusters is shown below.

```python
plt.plot(clusters, k_score_std, label="Standardized K-score")
plt.plot(clusters[elbow], k_score_std[elbow], "o", label="Elbow point")
plt.plot(clusters, dists_to_line, label="Distance from line segment")
plt.title('K score (standardized) with Elbow')
plt.xlabel('Number of clusters')
plt.ylabel('K score (standardized)')
plt.legend()
plt.show()

plt.plot(clusters, silhouettes, label="Silhouette score")
plt.plot(clusters[elbow], silhouettes[elbow], "o", label="Elbow point")
plt.title('Silhouette score with Elbow')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.legend()
plt.show()
```

## Comparing Users of the Same Label

First, a function to calculate the affinity of 2 MAL users is defined. The affinity of 2 users is synonymous to the Pearson's R coefficient of the ratings of anime they both watched.

```python
from scipy import stats

def affinity(uid1, uid2, rating_fv):
    user1_rating = rating_fv[uid1].toarray()[0]
    user2_rating = rating_fv[uid2].toarray()[0]
    to_remove1 = np.where(user1_rating == 0)
    to_remove2 = np.where(user2_rating == 0)
    to_remove = np.append(to_remove1, to_remove2)
    to_remove = np.unique(to_remove)
    user1_common_rating = np.delete(user1_rating, to_remove)
    user2_common_rating = np.delete(user2_rating, to_remove)
    affinity = stats.pearsonr(user1_common_rating, user2_common_rating)
    return affinity[0]
```

## Discussion

 ### Caveats

One caveat that should be noted about the project is that it is assumed that if a user has not rated an anime, even if they have watched the show, it is considered unwatched, since we cannot make assumptions about whether or not they liked the anime.

Another caveat is that rating scales vary per user. Some make full use of the 1-10 scale, while others tend towards a 6-10 scale, which may possibly be modeled after passing-failing grades in school. We ignored this fact of the matter, however future iterations of the project may use something like an exponential scale to give more weight to higher scores.

### Process analysis

As mentioned before, there exists better methods to cluster such a sparse and high-dimensional space, such as subspace clustering, however, these methods are not available in scikit-learn, and so were disregarded for the meantime. However, these methods may be recommended to future developers who wish to tackle a similar problem.

Metrics for choosing the number of components (200) were not evaluated, however a cumulative explained variance of 55% seems reasonable, given that no 2 users can have the exact same taste in anime (well, to some extent). That being said, the effectiveness of the use of a clustering method based on the Euclidean distance on a 200-dimension space is debatable, so for future iterations of this project, it is recommended that either a more reasonable dimension or a better distance metric (eg. cosine distance) is chosen. Note that the distance metric of the KMeans class in scikit-learn cannot be changed.

### Clustering analysis

While some users of the same label share similar tastes, some pairs can have low affinity scores, as opposed to pairs from different clusters. This could be due to the fact that the methodology favors pairs with more shared anime, regardless of their scores and affinity. For future iterations of the project, a clustering method that uses Pearson's R coefficient (or something similar) as a criterion may be used.
